```python
import pandas as pd
```

```python
df = pd.read_excel("JBL_MAR24_CLUSTERS_15MIN.xls")
```

```python
df.head()
```

```python
df['previous_state'] = df['Cluster'].shift(1)
```

```python
df.head()
```

```python
# drop first row
df1=df.iloc[1:, :]
```

```python
df1.head()
```

```python
# converting 'previous_state' from float to int
df1['previous_state'] = df1['previous_state'].astype(int)
```

```python
#df1.head()
```

```python
df1['c_states'] = df1['previous_state'].astype(str) + df1['Cluster'].astype(str)
```

```python
df1.head(100)
```

```python
df1['count'] = 1
df2 = df1.groupby(['c_states']).count()['count']
```

```python
# converting to df
df3 = df2.to_frame()
```

```python
df3.shape
```

```python
df3.head()
```

```python
for col in df3.columns:
    print(col)
```

```python
df3['percent'] = (df3['count'] / df3['count'].sum()) * 100
```

```python
df3['percent'] = df3['percent'].apply(lambda x: round(x, 1))
```

```python
df3
```

```python
df3.to_excel('Markov.xls', index=True)
```
