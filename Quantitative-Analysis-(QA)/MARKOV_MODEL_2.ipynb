```python
import pandas as pd
```

```python
df = pd.read_excel("IR Dec25_Cluster_Analysis.xlsx")
```

```python
df.head()
```

**previous state:**

```python
df['previous_state'] = df['Cluster'].shift(1)
```

```python
df.head()
```

```python
# drop first row
df1=df.iloc[1:, :]
```

```python
df1.head()
```

```python
# converting 'previous_state' from float to int
df1['previous_state'] = df1['previous_state'].astype(int)
```

```python
#df1.head()
```

**creating transition states:**

```python
df1['c_states'] = df1['previous_state'].astype(str) + df1['Cluster'].astype(str)
```

```python
df1.head(100)
```

```python
df2 = df1.groupby('c_states').size().reset_index(name = 'c_states_count')
df2.set_index('c_states', inplace = True)
df2
```

```python
df2['percent'] = (df2['c_states_count'] / df2['c_states_count'].sum())
```

```python
states = df2['percent'].reset_index()
```

```python
states
```

```python
states.to_excel('states.xlsx', index=True)
```

```python
# Extract start and end states from `c_states`
states["start_state"] = states["c_states"].str[0].astype(int)
states["end_state"] = states["c_states"].str[1].astype(int)
```

```python
states
```

**Markov transition matrix:**

```python
# Create the Markov transition matrix
transition_matrix = states.pivot(index="start_state", columns="end_state", values="percent").fillna(0)
```

```python
transition_matrix
```

```python
print(transition_matrix.sum(axis=1))
```

```python
# Convert the transition matrix values to percentages
transition_matrix_percent = (transition_matrix * 100).round(2)
transition_matrix_percent
```

**Applications:**

Prediction: Predict the next state of the market based on the current state (i.e., forecasting the market's behavior).

Optimization: Use the Markov Model for portfolio optimization, where states represent different market conditions, and transitions model how the market shifts.

Risk Assessment: Assess the risk of being in a certain state at a future time.
