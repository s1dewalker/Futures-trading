```python
import pandas as pd
```

```python
df = pd.read_excel(r'TRADE JOURNAL shared.xlsx')
df.head()
```

changing the headers

```python
# Set the first row as the new header
df.columns = df.iloc[0]

# Drop the first two rows (original row 0 and the new header row 1)
df = df.drop([0, 1]).reset_index(drop=True)
```

```python
# Set the display option to show all columns
pd.set_option('display.max_columns', None)
```

```python
df.head(10)
```

```python
pd.reset_option('display.max_columns')
```

checking the data types

```python
df.info(verbose=True)
```

converting data type to date time

```python
df['DATE2'] = pd.to_datetime(df['DATE2'])
```

```python
print(df['DATE2'].dtype)
```

```python
df.head()
```

```python
# Extract the day from the 'DATE2' column
df['Day'] = df['DATE2'].dt.day
```

```python
import numpy as np
df['Day_Category'] = np.where(df['Day'] <= 15, 1, 2)
```

converting data type to numeric time

```python
# Ensure 'TOTAL PROFIT' is numeric
df['TOTAL PROFIT'] = pd.to_numeric(df['TOTAL PROFIT'], errors='coerce')
df['PNL (PRICE)'] = pd.to_numeric(df['PNL (PRICE)'], errors='coerce')
df['E QTY'] = pd.to_numeric(df['E QTY'], errors='coerce')
```

```python
grouped_df = df.groupby('Day_Category')['TOTAL PROFIT'].sum().reset_index()
grouped_df
```

```python
df['TOTAL PROFIT'].sum()
```

```python
count_df = df.groupby('Day_Category')['W/L'].value_counts().unstack(fill_value=0).reset_index()
count_df
```

```python
# Reset the index for better readability
count_df = count_df.set_index('Day_Category')

count_df
```

```python
# Convert counts to percentages
percentage_df = count_df.div(count_df.sum(axis=1), axis=0)*100 #sum across the row and divide columnwise by that sum
percentage_df
```

```python
pnl_single = df.groupby('W/L')['PNL (PRICE)'].mean().reset_index()
pnl_single
```

```python
lot_sizing = df.groupby('W/L')['E QTY'].mean().reset_index()
lot_sizing
```

```python
# Extract the month from the 'DATE2' column
df['Month'] = df['DATE2'].dt.month
```

```python
lot_sizing = df.groupby(['Month','W/L'])['E QTY'].mean().reset_index()
lot_sizing
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
m_pnl = df.groupby(['Month','W/L'])['TOTAL PROFIT'].sum().reset_index()
m_pnl
```

```python
monthly_df=m_pnl.merge(lot_sizing,how='inner')
monthly_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filtered_df = monthly_df[monthly_df['W/L'].isin(['W', 'L'])]
monthly_df.head()
```

```python
# Define a color palette
color_palette = {
    'W': 'green',  # Set the color for Wins
    'L': 'red',    # Set the color for Losses
}
```

```python
plt.figure(figsize=(12, 8)) # figsize needs to be first
sns.barplot(data=filtered_df, x='Month', y='E QTY', hue='W/L', dodge = True, palette=color_palette)


# Add labels and title
plt.title('E QTY for Wins and Losses by Month')
plt.xlabel('Month')
plt.ylabel('E_QTY')


# Show the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=filtered_df, x='Month', y='E QTY', hue='W/L', marker='o',ci=None, palette=color_palette)
plt.title('QTY Trends for Wins and Losses by Month')
plt.xlabel('Month')
plt.ylabel('QTY')
plt.legend(title='Result')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_df, x='W/L', y='E QTY', palette=color_palette)
plt.title('Distribution of QTY for Wins and Losses')
plt.xlabel('Result')
plt.ylabel('E_QTY')
plt.show()
```

```python

plt.figure(figsize=(14, 8))
sns.lineplot(data=filtered_df, x='Month', y=filtered_df['TOTAL PROFIT'].abs(), hue='W/L', marker='o',ci=None, palette=color_palette)
plt.title('pnl Trends for Wins and Losses by Month')
plt.xlabel('Month')
plt.ylabel('TOTAL PROFIT')
plt.legend(title='Result')
plt.show()
```

```python
ticks = df.groupby(['Month','W/L'])['PNL (PRICE)'].mean().reset_index()
ticks
```

```python
monthly_df2=monthly_df.merge(ticks,how='inner')
monthly_df2
```

```python
filtered_df2 = monthly_df2[monthly_df2['W/L'].isin(['W', 'L'])]
filtered_df2.head(16)
```

```python
plt.figure(figsize=(16, 8)) # figsize needs to be first
sns.barplot(data=filtered_df2, x='Month', y=filtered_df2['PNL (PRICE)'].abs(), hue='W/L', dodge = True, palette=color_palette)


# Add labels and title
plt.title('PNL (PRICE) for Wins and Losses by Month')
plt.xlabel('Month')
plt.ylabel('PNL (PRICE)')



# Show the plot
#plt.tight_layout()
plt.show()
```
